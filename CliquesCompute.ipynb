```python
#load graphs
coauthorNetwork = CoauthorNetwork.load_with_loader("./data/outputacm.txt", parse_dataset_file);
```

```python
#compute cliques distribution
cliques_distribution = {}
for clique in list(nx.find_cliques(coauthorNetwork.gr)):
    cliques_distribution[len(clique)] = cliques_distribution.get(len(clique), 0) + 1

cliques_distribution_pairs = [(size, quantity) for size, quantity in cliques_distribution.items()]
print(cliques_distribution_pairs)
```
